```python
# Importanweisungen

import numpy as np
import statistics as stat
import scipy as sci
import sympy as sym
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.axes as axes
from matplotlib import colors as mcolors
import math
from scipy import optimize
import uncertainties as unc
import uncertainties.unumpy as unp 
import uncertainties.umath as umath
unv=unp.nominal_values
usd=unp.std_devs
```

```python
# Konstanten fuer einheitliche Darstellung

fig_size = (10, 6)
fig_legendsize = 14
fig_labelsize = 12
matplotlib.rcParams.update({'font.size': fig_labelsize})

colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
```

```python
# mathe Funktionen

def mean(n):
    # find the mean value and add uncertainties
    k = np.mean(n)
    err = stat.variance(unv(n))
    return unc.ufloat(unv(k), math.sqrt(usd(k)**2 + err))
```

```python
# allgemeine Fitfunktionen

def linear(x,m): # lineare Funktion mit f(x) = m * x 
    return(m*x)

def gerade(x, m, b): # gerade mit = f(x) = m * x + b
    return (m*x + b)

def cyclic(x, a, f, phi):
    return a * np.sin(x * f - phi)

def cyclicOff(x, a, f, phi, offset):
    return cyclic(x, a, f, phi) + offset

def gauss(x, x0, A, d, y0):
    return A * np.exp(-(x - x0)**2 / 2 / d**2) + y0

def exponential(x, c, y0):
    return np.exp(c * x) * y0

# fittet ein dataset mit gegebenen x und y werten, eine funktion und ggf. anfangswerten und y-Fehler
# gibt die passenden parameter der funktion, sowie dessen unsicherheiten zurueck
#
# https://stackoverflow.com/questions/14581358/getting-standard-errors-on-fitted-parameters-using-the-optimize-leastsq-method-i#
# Updated on 4/6/2016
# User: https://stackoverflow.com/users/1476240/pedro-m-duarte
def fit_curvefit(datax, datay, function, p0=None, yerr=None, **kwargs):
    pfit, pcov = \
         optimize.curve_fit(function,datax,datay,p0=p0,\
                            sigma=yerr, epsfcn=0.0001, **kwargs)
    error = [] 
    for i in range(len(pfit)):
        try:
          error.append(np.absolute(pcov[i][i])**0.5)
        except:
          error.append( 0.00 )
    pfit_curvefit = pfit
    perr_curvefit = np.array(error)
    return pfit_curvefit, perr_curvefit 

# usage zB:
# pfit, perr = fit_curvefit(unv(xdata), unv(ydata), gerade, yerr = usd(ydata), p0 = [1, 0])
# fuer eine gerade mit anfangswerten m = 1, b = 0
```

```python
# gibt Daten fuer LaTeX bereitgestellt aus

def latex_printValue(val, file):
    if isinstance(val, (list, tuple)):
        i = 1
        for v in val:
            latex_printValue(v, file+"_"+str(i))
            i += 1
    else:
        f = open('data/%s.txt' % file, 'w')
        f.write(str(val).replace('/', ''))
```

```python
# weitere Werte, Konstanten
# Werte von https://physics.nist.gov/cuu/Constants/index.html[0]

c = 299792458 # m/s
k_B = unc.ufloat_fromstr("1.38064852(79)e-23") # J K-1 [0]
h = unc.ufloat_fromstr("4.135667662(25)e-15") # eV s [0]
r_e = unc.ufloat_fromstr("2.8179403227(19)e-15") # m [0]
R = unc.ufloat_fromstr("8.3144598(48)") # J mol-1 K-1 [0]
K = 273.15 # kelvin 
g = 9.81 # m/s^2
rad = 360 / 2 / math.pi
grad = 1/rad
```

```python
# import der messwerte

unc_x = 0.1 / 2 / np.sqrt(6) # Messleiste; analog in cm
unc_U = 0.01 / 2 / np.sqrt(3) # Voltmeter; digital in Volt(20V)
unc_phi = 1 / 2 / np.sqrt(6) # Winkel; analog in Grad

def converter(data):
    try:
        return float(s if len(s) > 0 else 0)
    except:
        return data
        
con = {}
for i in range(50):
    con[i] = converter
    
x_add = unc.ufloat(42.5, unc_x)
x_halterung = unc.ufloat(2.9, unc_x)/2

profil = []

for i in range(1,5):
    data = np.loadtxt("messung/profil%i.csv" % i, skiprows = 2, delimiter = ",")
    x = unp.uarray(data[:,0], unc_x) + x_add + x_halterung
    y = unp.uarray(data[:,1], unc_x) - unp.uarray(data[:,2], unc_x)
    I = unp.uarray(data[:,3], unc_U)
    profil.append((x, y, I))
#print(profil)

data = np.loadtxt("messung/stehendeWelle.csv", skiprows = 1, delimiter = ",")
stehend_min = unp.uarray(data[:], unc_x)

data = np.loadtxt("messung/pvc_single.csv", skiprows = 1, delimiter = ",")
p1 = unp.uarray(data[:,0], unc_phi) * grad
p2 = unp.uarray(data[:,1], unc_phi) * grad
pvc = (45*grad - p1, p2 - p1 - 135*grad)

data = np.loadtxt("messung/pvc_double.csv", skiprows = 1, delimiter = ",")
y = unp.uarray(data[:,0], unc_x) - unp.uarray(data[:,1], unc_x)
I = unp.uarray(data[:,2], unc_U)
pvc2 = (y, I)

data = np.loadtxt("messung/schaumstoff.csv", skiprows = 1, delimiter = ",")
p1 = unp.uarray(data[:,0], unc_phi)*grad
I = unp.uarray(data[:,1], unc_U)
schaumstoff = ((360*grad-p1)/2, I)
```

```python
# Rechnung profil

p = []
for (x, y, I) in profil:    
    xdata, ydata = y, I
    pfit, perr = fit_curvefit(unv(xdata), unv(ydata), gauss, yerr = usd(ydata), p0 = [0, 1, 10, 0])
    p.append(unp.uarray(pfit, perr))
p = np.abs(p)

hwb = 2 * math.sqrt(2 * math.log(2)) * p[:,2]

#print(profil)
xdata, ydata = [i[0][0] for i in profil], hwb/2 + p[:,0]
pfit, perr = fit_curvefit(unv(xdata), unv(ydata), gerade, yerr = usd(ydata), p0 = [1,0])
l1 = unp.uarray(pfit, perr)
xdata, ydata = [i[0][0] for i in profil], -hwb/2 + p[:,0]
pfit, perr = fit_curvefit(unv(xdata), unv(ydata), gerade, yerr = usd(ydata), p0 = [1,0])
l2 = unp.uarray(pfit, perr)

x_quell = (l2[1] - l1[1]) / (l1[0] - l2[0])
y_quell1 = gerade(x_quell, *l1)
y_quell2 = gerade(x_quell, *l2)
quell = (x_quell, (y_quell1 + y_quell2)/2)
quell
latex_printValue(quell, "quellfleck_cm")

phi = umath.atan(l1[0]) - umath.atan(l2[0])
phi / 2 / math.pi * 360 # grad
latex_printValue(phi, "phiProfil_rad")
latex_printValue(phi / 2 / math.pi * 360, "phiProfil_grad")
```

```python
# Diagramm Strahlprofil

fig=plt.figure(figsize=fig_size)
ax = plt.gca()

i = 1
for (x, y, I) in profil:
    
    color = next(ax._get_lines.prop_cycler)['color']
    
    xdata, ydata = x+I, y
    plt.plot(unv(x),unv(y), linewidth=1, color = color)
    plt.plot(unv(x[0]) + gauss(np.arange(-30,30), *unv(p[i-1])), np.arange(-30, 30), linewidth=1, color = color)
    plt.errorbar(unv(xdata),unv(ydata),usd(ydata),usd(xdata),fmt=' ', capsize=5, linewidth=1, color = color, label='Profil %i'%i)
    
    i += 1

color = next(ax._get_lines.prop_cycler)['color']
xdata, ydata = [i[0][0] for i in profil], hwb/2 + p[:,0]
plt.errorbar(unv(xdata),unv(ydata),usd(ydata),usd(xdata),fmt=' ', capsize=10, linewidth=1, color = color)
xdata, ydata = [i[0][0] for i in profil], -hwb/2 + p[:,0]
plt.errorbar(unv(xdata),unv(ydata),usd(ydata),usd(xdata),fmt=' ', capsize=10, linewidth=1, color = color)

xrange = np.linspace(-10,130)
plt.plot(xrange, gerade(xrange, *unv(l1)), linewidth = 2, color = color, label = "Aufweitung")
plt.plot(xrange, gerade(xrange, *unv(l2)), linewidth = 2, color = color)
plt.annotate('$\phi$', xy = unv(quell), xytext=unv(quell) + (10,-0.5))
    

# plt.xlim(400,401)
plt.legend(prop={'size':fig_legendsize})
plt.grid()
plt.tick_params(labelsize=fig_labelsize)
plt.xlabel("Position des Empfängers $x$ $t$ [$cm$]", {'fontsize':fig_legendsize+2})
plt.ylabel("Position des Empfängers $y$ [$cm$]", {'fontsize': fig_legendsize+2})
plt.savefig("data/profil.pdf")
plt.show()
```

```python
# rechnung pvc 
# snellius n1 sin phi1 = n2 sin phi2; mit n2 = 1, n1 = n
# n sin 1 = sin 2 <=> n = sin 2 / sin 1

(phi1, phi2) = pvc
n = unp.sin(phi2)/unp.sin(phi1)
latex_printValue(mean(n), "n_pvc_ungenau")
```
