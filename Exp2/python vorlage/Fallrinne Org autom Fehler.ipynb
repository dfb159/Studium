```python
# Importanweisungen

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.axes as axes
import math
from scipy import optimize
import uncertainties as uc
import uncertainties.unumpy as unp 
unv=unp.nominal_values
usd=unp.std_devs
```

```python
# Konstanten fuer einheitliche Darstellung

fig_size = (10, 6)
fig_legendsize = 14
fig_labelsize = 12
```

```python
# allgemeine Fitfunktionen

def linear(x,m): # lineare Funktion mit f(x) = m * x 
    return(m*x)

def gerade(x, m, b): # gerade mit = f(x) = m * x + b
    return (m*x + b)

# fittet ein dataset mit gegebenen x und y werten, eine funktion und ggf. anfangswerten und y-Fehler
# gibt die passenden parameter der funktion, sowie dessen unsicherheiten zurueck
#
# https://stackoverflow.com/questions/14581358/getting-standard-errors-on-fitted-parameters-using-the-optimize-leastsq-method-i#
# Updated on 4/6/2016
# User: https://stackoverflow.com/users/1476240/pedro-m-duarte
def fit_curvefit(datax, datay, function, p0=None, yerr=None, **kwargs):

    pfit, pcov = \
         optimize.curve_fit(function,datax,datay,p0=p0,\
                            sigma=yerr, epsfcn=0.0001, **kwargs)
    error = [] 
    for i in range(len(pfit)):
        try:
          error.append(np.absolute(pcov[i][i])**0.5)
        except:
          error.append( 0.00 )
    pfit_curvefit = pfit
    perr_curvefit = np.array(error)
    return pfit_curvefit, perr_curvefit 

# usage zB:
# pfit, perr = fit_curvefit(unv(xdata), unv(ydata), gerade, yerr = usd(ydata), p0 = [1, 0])
# fuer eine gerade mit anfangswerten m = 1, b = 0
```

```python
# gibt Daten fuer LaTeX bereitgestellt aus

def latex_printValue(val, file):
    f = open('data/%s.txt' % file, 'w')
    f.write(str(val).replace('/', ''))
```

```python
# import der messwerte

data = np.loadtxt("rinne.txt", comments = "#", delimiter = "\t", unpack=False)
s = unp.uarray(data[:,0], 1e-3/np.sqrt(6)) # abstand zu oben in erster spalte mit unsicherheit 1mm
a = unp.uarray(data[:,1], 1e-2/np.sqrt(6)) # auslenkung in zweiter spalte mit unsicherheit 1cm
a # auslenkungen
```

```python
#geometrie Fallrinne und relevante hoehe

L = uc.ufloat(0.5, 3e-3 / np.sqrt(6)) # laenge nach skizze
H = uc.ufloat(0.327, 3e-3 / np.sqrt(6)) # gesammthoehe
h0 = H - 0.025 # maximale fallhoehe
h = h0 - s / unp.sqrt(1 + L**2 / H**2) # fallhoehe der einzelnen kugellagen
h # fallhoehen
```

```python
# diagramm 1: auslenkung gegen wurzel der fallhoehe

sqrth = unp.sqrt(h) # wurzel aus fallhöehe

fig=plt.figure(figsize=fig_size)

p0 = None # startwerte, wenn unsicher: p0 = None
pfit, perr = fit_curvefit(unv(sqrth), unv(a), gerade, yerr = usd(a))
p = unp.uarray(pfit, perr)

# gibt die Parameter in extra files aus
latex_printValue(p[0], 'rutsche_steigung')
latex_printValue(p[1], 'rutsche_achsenabschnitt')

plt.plot(unv(sqrth), gerade(unv(sqrth), *pfit), label='Linearer Fit:\nm = %s\ny0 = %s' % tuple(p))
plt.errorbar(unv(sqrth),unv(a),usd(sqrth),usd(a),fmt='none', capsize=5, linewidth=2, label='Messwerte mit Fehlerbalken')

#plt.xlim(0,1200)
plt.legend(prop={'size':fig_legendsize})
plt.grid()
plt.tick_params(labelsize=fig_labelsize)
plt.xlabel("Wurzel der Kugelhöhe über dem Kollisionspunkt [$\sqrt{m}$]", {'fontsize':fig_legendsize+2})
plt.ylabel("Auslenkung des Pendels [m]", {'fontsize': fig_legendsize+2})
plt.savefig("data/Rinne.pdf")
plt.show()
```
