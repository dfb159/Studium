```python
import numpy as np
import scipy as sp
import sympy as smp
from IPython.display import display, Math, Latex
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.axes as axes
import math
import statistics as stat
from scipy import optimize
```

```python
w=np.array([2000,2200,2500,2700,3000])*2*np.pi/60
uf=40/(2*np.sqrt(3))
uw=uf*2*np.pi/60
T_2=np.array([44.97/5,29.5/3,33.57/3,24.65/2,27.22/2])
uT_2=np.sqrt((0.4/(2*np.sqrt(3)))**2 + (0.01/(2*np.sqrt(6)))**2)
T=(T_2)
uT=np.array([(uT_2)/5,(uT_2)/3,(uT_2)/3,(uT_2)/2,(uT_2)/2])
```

```python
def f(w,m):
    return (w*m)

fig=plt.figure(figsize=(10,6))
p0=[0.1]
popt, pcov, infodict, errmsg, ier = optimize.curve_fit(f,w,T,p0=p0, full_output= True)
plt.plot(w,f(w,*popt),label='Ausgleichskurve')
plt.errorbar(w,T,uT,uw,fmt='none', capsize=5, linewidth=2, label='Messwerte mit Fehlerbalken')



plt.legend(prop={'size':14})
plt.grid()
plt.tick_params(labelsize=22)
plt.ylabel("Präzessionszeit $T_p$ in s", {'fontsize':22})
plt.xlabel("Winkelgeschwindigkeit $\omega$ in 1/s", {'fontsize': 22})
plt.savefig("sproHzmitte.pdf")
plt.show()
```
